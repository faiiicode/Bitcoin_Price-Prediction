# BITCOIN PRICE PREDICTION



```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
df=pd.read_csv('BitcoinPrice (3).csv')
```

```python
df.head(30)
```

```python
df.info()
```

```python
df.drop(['Date'],1,inplace=True)
```

```python
df.head()
```

```python
#predecting 30 days in future
```

```python
p_days=30
df['Prediction']=df[['Price']].shift(-p_days)
                          
```

```python
df.head()
```

```python
#seperate X and Y
```

```python
X=np.array(df.drop(['Prediction'],1))
X=X[: len(df)- p_days]
```

```python
X.shape
```

```python
X
```

```python
Y=np.array(df['Prediction'])
Y=Y[: -p_days]
```

```python
Y.shape
```

```python
from sklearn.model_selection  import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
P_days=np.array(df.drop(['Prediction'], 1))[-p_days :]
```

```python
print(P_days)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
Rf=RandomForestRegressor(n_estimators=1000, random_state=1)
```

```python
Rf.fit(X_train,Y_train)
```

```python
print('Radom Forest Accuracy : {:.2f}%'.format(Rf.score(X_test,Y_test)*100))
```

```python
#Prediction
```

```python
Rf_PRed=Rf.predict(X_test)
```

```python
print(Rf_PRed)
```

```python
Rf_PRed_30=Rf.predict(P_days)
print(Rf_PRed)
```
